```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
product_name =[]
price =[]
```

```python
for i in range(1,21):
    url = 'https://www.amazon.in/s?k=laptop+under+200000&crid=2GC3DM8LELY65&qid=1687634726&sprefix=laptop+under+200000%2Caps%2C340&ref=sr_pg_'+str(i)

    r = requests.get(url)

    soup = BeautifulSoup(r.text,'lxml') 

    names = soup.find_all('h2',class_ = 'a-size-mini a-spacing-none a-color-base s-line-clamp-2')
    for i in names:
        product_name.append(i.text)

    cost = soup.find_all('span',class_ = 'a-price-whole')
    for i in cost:
        price.append(i.text)
```

```python
print(product_name)
print(price)
```

```python
data = pd.DataFrame({'Product_Name':product_name,'Price':price})
print(data.head())
```

```python
data.shape
```

```python
data['Description'] = data['Product_Name'].str.split(' ').str[3:].str.join(' ')
```
